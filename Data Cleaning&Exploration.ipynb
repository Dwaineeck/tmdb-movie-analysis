```python
import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical operations
import matplotlib.pyplot as plt  # For plotting graphs
import seaborn as sns  # For statistical data visualization
```

```python
#LoadDataset
df = pd.read_csv(r'D:\thenumerix\Data Cleaning & Exploration\TMDB_movie_dataset_v11.csv')
```

```python
#ExploreData
df.head()  # View the first 5 rows of the dataset
```

```python
df.info()  # Overview of the dataset, including data types and non-null counts
df.describe()  # Summary statistics for numerical columns
```

```python
df.isnull().sum()  # Count of missing values in each column
```

```python
df['budget'] = df['budget'].fillna(df['budget'].median())  # Fill missing budget values with the median
df = df.dropna(subset=['revenue'])  # Drop rows where the revenue column is missing
```

```python
#VISUALIZE KEY STATISTICS#
```

```python
sns.histplot(df['budget'], bins=20)  # Histogram of the budget column
plt.show()  # Display the plot
```

```python
#CORRELATION HEATMAP#
```

```python
numeric_df = df.select_dtypes(include=[np.number])  # Select only numeric columns
plt.figure(figsize=(10, 8))  # Set the figure size
sns.heatmap(numeric_df.corr(), annot=True)  # Heatmap of correlation matrix with annotations
plt.show()  # Display the plot
```
